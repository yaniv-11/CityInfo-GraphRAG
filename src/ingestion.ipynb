```python
from pdf2image import convert_from_path
import pytesseract
import os

# Convert PDF to images
images = convert_from_path(pdf_path)

ocr_texts = []
for i, image in enumerate(images):
    try:
        # Perform OCR on each image
        text = pytesseract.image_to_string(image)
        if text:
            ocr_texts.append({"page": i + 1, "content": text.strip(), "type": "text"})
    except Exception as e:
        print(f"An error occurred during OCR on page {i+1}: {e}")

print(f"✅ Extracted text from {len(ocr_texts)} pages using OCR")
if ocr_texts:
    display(ocr_texts[:2])  # show first few entries
else:
    print("No text extracted using OCR.")
```

```python
import re

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)                     # Normalize whitespace
    text = re.sub(r'[^a-zA-Z0-9.,:;!?()/%\- ]+', '', text)  # Remove junk
    return text.strip()

def is_meaningful(text):
    if len(text.split()) < 3:
        return False
    if re.fullmatch(r'[^a-zA-Z0-9]+', text):
        return False
    return True

cleaned_texts = []

for item in ocr_texts:
    # 🧠 Handle both dict and string cases
    if isinstance(item, dict):
        text = item.get('content', '')
    elif isinstance(item, str):
        text = item
    else:
        continue

    t_clean = clean_text(text)
    if is_meaningful(t_clean):
        cleaned_texts.append(t_clean)

print(f"✅ Kept {len(cleaned_texts)} / {len(ocr_texts)} meaningful chunks")

```

```python
for i, chunk in enumerate(cleaned_texts[:3]):  # show first 3 pages
    print(f"\n---- Page {i+1} ----\n")
    print(chunk[:1000])  # print first 1000 chars to avoid overload
```

```python
from sentence_transformers import SentenceTransformer

# choose a strong general-purpose embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(cleaned_texts, show_progress_bar=True)
print("✅ Generated embeddings shape:", embeddings.shape)
```

```python
import pickle
import os

output_dir = "/content/embeddings"
os.makedirs(output_dir, exist_ok=True)
embeddings_path = os.path.join(output_dir, "ocr_embeddings.pkl")

with open(embeddings_path, "wb") as f:
    pickle.dump(embeddings, f)

print(f"✅ Embeddings saved to {embeddings_path}")
```

To use the embeddings for retrieval, you'll need a query. Let's define a sample query and generate its embedding.

```python
# Define a sample query
query = "What is the economic profile of Bengaluru Urban?"

# Generate embedding for the query
query_embedding = model.encode(query)

print("✅ Generated query embedding shape:", query_embedding.shape)
```

Now, you can calculate the similarity between the query embedding and all the document embeddings. Cosine similarity is a common metric for this.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between the query embedding and all document embeddings
similarities = cosine_similarity([query_embedding], embeddings)

# The result is a 2D array, so we take the first row
similarities = similarities[0]

print("✅ Calculated similarities.")
```

Finally, you can retrieve the top N most similar document chunks based on the similarity scores.

```python
import numpy as np

# Get the indices of the top N most similar chunks
top_n = 3  # You can adjust this number
top_indices = np.argsort(similarities)[::-1][:top_n]

print(f"✅ Top {top_n} most similar chunks:")

# Display the top chunks and their similarity scores
for i in top_indices:
    print(f"\n---- Page {ocr_texts[i]['page']} (Similarity: {similarities[i]:.4f}) ----\n")
    print(ocr_texts[i]['content'][:500] + "...") # Print first 500 characters of content
```
